```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('results.csv')
```

```python
plt.figure(figsize=(9,9))
plt.plot(df['points'], df['Global_4'], label='Global order=4')
plt.plot(df['points'], df['Block_4'], label='Block order=4')
plt.legend()
plt.xlabel('grid points (mega points)')
plt.ylabel('bandwith Gbyte/s')
plt.show()

plt.figure(figsize=(9,9))
plt.plot(df['points'], df['Block_2'], label='Block order=2')
plt.plot(df['points'], df['Block_4'], label='Block order=4')
plt.plot(df['points'], df['Block_8'], label='Block order=8')
plt.legend()
plt.xlabel('grid points (mega points)')
plt.ylabel('bandwith Gbyte/s')
plt.show()
```

```python
fn = '2000.csv'

d = np.array([float(s) for s in open(fn,'r').readlines()[-1].replace('\n','').split(',')[:-1]])

N = int(np.sqrt(d.shape[0]))
x = np.arange(N)
X,Y = np.meshgrid(x,x)
d = d.reshape((N,N))

plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X,Y, d, rstride=1, cstride=1,
                cmap='jet', edgecolor='none')
ax.set_title('surface');
plt.colorbar(surf)
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlim(-1,1)
plt.show()
plt.close()
```
